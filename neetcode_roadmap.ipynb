# Arrays & Hashing

## 1. Two Sum 

```python
from typing import List

class Solution:
    def twoSum(self, nums: List[int], target: int) -> List[int]:
        diff = {}
        for i in range(len(nums)):
            if nums[i] in diff:
                return diff[nums[i]], i
            diff[target - nums[i]] = i

ob = Solution()
ob.twoSum([3, 3], 6)
```
